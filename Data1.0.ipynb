```python
import pandas as pan
import matplotlib.pyplot as plot
```

```python
b_data = pan.read_csv("data/boston_housing_data.csv", sep=',')
```

```python
NE_suburb = b_data.loc[b_data['age'] == b_data.age.min()]
NE_suburb.medv.mean() * 1000
```

```python
crime = b_data.crim
teachers = b_data.ptratio
river = b_data.chas
color = ['red' if r == 0 else 'blue' for r in river]
```

```python
plot.scatter(teachers, crime, c=color)
plot.title('Per-capita crime rate and the pupil-teacher ratio Blue = bounded by the Charles River \n')
plot.xlabel("Pupil-teacher ratio")
plot.ylabel("Per capita crime rates")
plot.show()
```

```python
black_citizens = b_data.black/10
distance = b_data.dis
```

```python
plot.scatter(black_citizens, distance)
plot.title('% of black citizens and the avg dis to employment centers in Boston by suburb \n')
plot.xlabel("% of black citizens")
plot.ylabel("Weighted avg distance to employment centers")
plot.show()
```
